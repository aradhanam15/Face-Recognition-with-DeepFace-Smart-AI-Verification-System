```python
from google.colab import files
import os

# Create a folder to store known faces
os.makedirs("known_faces", exist_ok=True)

# Upload images of known people (e.g., aradhana.jpg, neha.jpg, alia.jpg)
uploaded = files.upload()

# Move uploaded files into known_faces folder
for file_name in uploaded.keys():
    os.rename(file_name, f"known_faces/{file_name}")
```

```python

from deepface import DeepFace
# Store known faces in a list
known_faces = []
known_names = []

# Loop through all files in known_faces
for file in os.listdir("known_faces"):
    if file.endswith(('.jpg', '.jpeg', '.png')):
        known_faces.append(f"known_faces/{file}")
        known_names.append(os.path.splitext(file)[0])  # 'aradhana', 'neha', etc.
```

```python
uploaded_test = files.upload()
test_image_path = list(uploaded_test.keys())[0]
```

```python
recognized_people = []

for known_face_path, known_name in zip(known_faces, known_names):
    try:
        result = DeepFace.verify(img1_path=known_face_path, img2_path=test_image_path, enforce_detection=False)
        if result["verified"]:
            print(f"✅ Recognized: {known_name}")
            recognized_people.append(known_name)
        else:
            print(f"❌ Not matched: {known_name}")
    except Exception as e:
        print(f"Error with {known_name}: {e}")
```

```python
#Face Recognition Attendance System (Log to CSV)
from datetime import datetime
import pandas as pd
```

```python
now = datetime.now()
date = now.strftime("%Y-%m-%d")
time = now.strftime("%I:%M %p")

attendance_data = []

for name in recognized_people:
    attendance_data.append({"Name": name, "Date": date, "Time": time})

# Save or append to CSV
df = pd.DataFrame(attendance_data)
csv_path = "attendance.csv"

if os.path.exists(csv_path):
    df.to_csv(csv_path, mode='a', header=False, index=False)
else:
    df.to_csv(csv_path, index=False)

print("✅ Attendance logged!")
```
